```python
#PDB -> Protein data bank (three dimennsional structural) including proteins, nucleic acids
import Bio
from Bio.PDB.PDBParser import PDBParser

parser = PDBParser()
```

```python
#Read PDB File
structure_PDB = parser.get_structure('6lu7', file='compbio_file/6lu7.pdb') #namanya bebas tapi biar ga bingung dibuat sama aja
print(structure_PDB) #jangan lupa masuk dulu di folder compbio_file karna 6lu7.pdb ada disana
```

```python
#Parse Information
model_PDB = structure_PDB[0]
print(model_PDB)
```

```python
#Chain
for chain in model_PDB:
    # print(f'Chain {chain.id}')
    print(f'Chain {chain}, ID {chain.id}')
```

```python
#Residue
for chain in model_PDB:
    print(f'Chain {chain}, ID {chain.id}')
    for residue in chain:
        print(residue)
```

```python
#Ambil atom
for chain in model_PDB:
    print(f'Chain {chain}, ID {chain.id}')
    for residue in chain:
        print(residue)
        for atom in residue:
            print(atom)
```

```python
#Read File FASTA
from Bio import SeqIO
```

```python
#Read File
# import pprint
fasta_parse = SeqIO.parse('compbio_file/coronavirus.fasta', 'fasta')

#Parse
for record in fasta_parse:
    print(record.id)
    print(record.name)
    print(record.description)
    print(record.features)
    print(record.seq)
```

```python
#Method Read
fasta_read = SeqIO.read('compbio_file/coronavirus.fasta', 'fasta')

#Parse
print(fasta_read.id)
print(fasta_read.description)
print(fasta_read.features)
print(fasta_read.seq)
```

```python
#GenBank
genbank_parse = SeqIO.parse('compbio_file/coronavirus.gb', 'gb')

#Parse
for record in genbank_parse:
    print(record.id)
    print(record.name)
    print(record.description)
    print(record.features)
    print(len(record.features))
    print(record.seq)
```

```python
#Read
genbank_read = SeqIO.read('compbio_file/coronavirus.gb', 'gb')
print(genbank_read.id)
print(genbank_read.name)
print(genbank_read.description)
print(genbank_read.features)
print(len(genbank_read.features))
print(len(genbank_read.seq))
```

```python
print("BEFORE SPLIT")
print(genbank_read.seq.translate())

#Pisahin *
split_seq = genbank_read.seq.translate().split("*")
print("AFTER SPLIT")
for protein in split_seq:
    print(f"Protein : {protein}, Length : {len(protein)}")
```

```python
#Split Sequence
Seq_Source = genbank_read.seq

#SeqA -> 0 - 1/3 Source
SeqA = Seq_Source[: round(len(Seq_Source)/3)]
print(len(SeqA))

#SeqB -> 1/3 - 2/3 Source
SeqB = Seq_Source[round(len(Seq_Source)/3) : round(len(Seq_Source)*2/3)] #
print(len(SeqB))
```

```python
#Import Melting Temperature
from Bio.SeqUtils import MeltingTemp as mt
```

```python
#Calculate MT for each Sequence
mt_SeqA = mt.Tm_GC(SeqA)
print(mt_SeqA)

mt_seqB = mt.Tm_GC(SeqB)
print(mt_seqB)
```

```python
#Bar Plot
%pip install matplotlib
import matplotlib.pyplot as plt
```

```python
#Plotting
plt.subplot(1, 2, 1,)
plt.bar(['SeqA', 'SeqB'], [mt_SeqA, mt_seqB])
plt.show()
```
