```python
#Creat DNA Sequence and calculate the length
import Bio
from Bio.Seq import Seq
```

```python
dna_seq = Seq('ATGATCTCGTAA')
```

```python
len(dna_seq)
```

```python
dna_seq
```

```python
#Calculate DNA Sequence Complement
#Be mind that AI = 2 hydrogen bonds GC = 3 hydrogen bonds
dna_seq.complement()
```

```python
#Calculate DNA Sequence reverse complement
dna_seq.reverse_complement()
```

```python
#costume function to calculate DNA sequence complement
dna_seq.complement()[::-1]
```

```python
dna_seq
```

```python
#Transcription : DNA to mRNA (writing the message )
dna_seq.transcribe()
```

```python
mRNA = dna_seq.transcribe()
```

```python
#Translation : mRNA to Protein/Amino Acid
#Method 1
mRNA.translate()
```

```python
#Method 2
#Direct translation of DNA to Amino Acid
dna_seq.translate()
```

```python
#Create our custome stop codon symbol
mRNA.translate(stop_symbol="@")
```

```python
#Back Transcription : mRNA to DNA
mRNA.back_transcribe()
```

image.png

```python
mRNA.back_transcribe() == dna_seq
```

```python
#join the steps
dna_seq.transcribe().translate()
```

```python
#Reading PDB files
from Bio.PDB.PDBParser import PDBParser
#Create a parser
parser = PDBParser()
structure = parser.get_structure("6LU7","6lu7.pdb")
```

```python
#The overall layout of a structure object follows rge so-called SMCRA
#SMCRA : Structure Model Chain Residu Atom architecture
structure
```

```python
#Model in the structure
model = structure[0]
```

```python
#Check for chains
for chain in model:
  print(f'Chain {chain},Chain_ID{chain.id}')
```

```python
#check for residu
for chain in model:
  print(f'Chain {chain},Chain_ID{chain.id}')
  for residu in chain:
    print(residu)
```

```python
#Check for atoms
for chain in model :
  print(f'Chain {chain},Chain_ID{chain.id}')
  for residu in chain:
    for atom in residu:
      print(atom)
```
