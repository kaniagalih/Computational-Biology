```python
from Bio import SeqIO

#FASTA
#1. Read 
fasta_seq = SeqIO.read('coronavirus.fasta', 'fasta')
print(fasta_seq.id) #kalau mau ambil spesifik (seq.id, seq.seq)

#2. Parse 
for record in SeqIO.parse('coronavirus.fasta', 'fasta'):
    fasta_rec = record.seq
print(fasta_rec)
```

```python
#GENBANK
#1. Read
genbank_seq = SeqIO.read('coronavirus.gb', 'genbank')
seq = genbank_seq.id
print(seq)

#2. Parse 
for record2 in SeqIO.parse('coronavirus.gb', 'genbank'):
    genbank_rec = record2.seq
print(genbank_rec)
```

SEQUENCE MANIPULATION 

```python
#[index awal : index akhir : lompatan antar index] 
#Print -1 -> reverse 
print(genbank_rec[::-1])
#Print dari index 0-8
print(genbank_rec[0:8:])
#Print dari index 0-8 dengan lompatan 3
print(genbank_rec[0:8:3])
#Print dari index 0-total seq/2
print(genbank_rec[:round(len(genbank_rec)/2)])
```

```python
seq1 = genbank_rec[:round(len(genbank_rec)/2)]
seq2 = genbank_rec[round(len(genbank_rec)/2):]
print('seq1 :', seq1)
print('seq2 :', seq2)
print('merge :', seq1 + seq2)

#reverse seq2 + merge with seq1
merge = seq2[::-1] + seq1
print('Merge with reverse :', merge)
```

```python
#Replace 
replace1 = seq1.replace('AT', 'GC')
print(seq1)
print(replace1)
```

SEQUENCE ANALYSIS

```python
SeqA = seq1[:26]
print(SeqA)
```

```python
#GC/AT CONTENT 
from Bio.SeqUtils import GC

gc = GC(SeqA)
print(gc)

at = 100 - gc
print(at)
# print('AT Content :' + str(at)) #sama sih trs beda nya ga pake str apa?
```

```python
#MELTING TEMP (derajat)
from Bio.SeqUtils import MeltingTemp as mt

#Melting Temp GC
meltingTemp_GC = mt.Tm_GC(SeqA)
print(meltingTemp_GC)

#Melting Temp Wallace 
meltingTemp_Wallace = mt.Tm_Wallace(SeqA)
print(meltingTemp_Wallace)

#Melting Temp NN
meltingTemp_NN = mt.Tm_NN(SeqA)
print(meltingTemp_NN)
```

```python
#MOLECULAR WEIGHT 
from Bio.SeqUtils import molecular_weight as mw

MolWeight = mw(SeqA)
print(MolWeight)
```

```python
#FINDING NUCLEOTIDES, CALCULATE NUCLEOTIDES FREQ
from collections import Counter

dna = Counter(SeqA)
key = dna.keys()
values = dna.values()

print("Keys :", key)
print("Values :", values)
print(dna.most_common())

plt.bar(key, values)
```

PROTEIN SYNTHESIS 
- Transcription -> DNA to RNA 
- Translation -> RNA to Amino Acid (polypeptide)


```python
#TRANSCRIPTION
#DNA : A-T G-C 
#RNA : A-U (antisense to RNA) T-U (sense to RNA)

#Compliment DNA
complement = SeqA.complement()

#DNA to RNA
mRNA = SeqA.transcribe()
print('DNA Sense : ', SeqA)
print('DNA AntiSense :', complement)
print('mRNA : ', mRNA)

#Back to DNA using RNA
dna = mRNA.back_transcribe()
print('Back to DNA :', dna)
```

```python
#TRANSLATION -> translation dr DNA ataupun mRNA (sama)
from Bio.SeqUtils import seq3

AminoAcid = mRNA.translate()
print('Amino Acid : ', AminoAcid)
print('Protein : ', seq3(AminoAcid))
```

SEQUENCE ALIGNMENT & SIMILARITIES

```python
#GLOBAL ALIGNMENT 
from Bio import pairwise2
from Bio.SeqUtils import Seq

SeqB = Seq('CGATCGATGCATGGCCTAACGTATA')
SeqC = Seq('CGATCGATGCAT')

#Global Alignment
  
global_align = pairwise2.align.globalxx(SeqA, SeqB)
global_align
```

```python
#SCORE FOR GLOBAL ALIGNMENT 
#Ditambahin (one_alignment_only=True, score_only=True) kalau mau print score aja
score = pairwise2.align.globalxx(SeqA, SeqB, one_alignment_only=True, score_only=True)
score
```

```python
#MAXIMUM SIMILARITY SCORE 
sama = 2
beda = -1
startGap = -0.5
extendGap = -0.1

#Ditambahin (one_alignment_only=True, score_only=True) kalau mau print score aja 
score_max = pairwise2.align.globalms(SeqA, SeqB, sama, beda, startGap, extendGap, one_alignment_only=True, score_only=True)
score_max
```

```python
#PERSENTASI SIMILARITY FOR MAX 
score / len(SeqA) * 100
```

```python
#PERSENTASE SIMILARITY FOR GLOBAL ALIGNMENT
global_align / len(SeqA) * 100
```

```python
#LOCAL ALIGNMENT 
local_align = pairwise2.align.localxx(SeqA, SeqC)
local_align
```

```python
#FORMAT ALIGNMENT 
from Bio.pairwise2 import format_alignment as fa

#Mau liat format alignment nya, dan emang butuh *
#Score 14 (seberapa banyak yang sama)
for alignmnent in global_align:
    print(fa(*alignmnent))
```

```python
#HAMING DISTANCE : Kasih tau berapa banyak leter yang beda

#Cara 1
haming1 = len([(SeqA, SeqB) for x,y in zip (SeqA, SeqB) if x !=y])
print(haming1)

#Cara 2 (pakai variable kosong, trs looping)
list1 = []
for x,y in zip(SeqA, SeqB):
    if x != y:
        list1.append((x,y))
# print(len(list1))
haming2 = len(list1)
print(haming2)
```

PLOTTING

```python
#DOT PLOT 
#Kalau seq nya sama maka print Black Square, kalo beda print blank
def dotplot(SeqA, SeqB):
    blank = ' ' #Blank
    fill = chr(0x25A0) #Black Square
    print(' |'+SeqA)
    for i in SeqB:
        print(i+'|', end='')
        for j in SeqA:
            if i == j: #kalo i=j maka print kotak 
                print(fill, end='') #Bakal print kotak 
            else:
                print(blank,end='') #Bakal print blank
        print() #Print ke baris baru
    
dotplot(SeqA, SeqB)
```
